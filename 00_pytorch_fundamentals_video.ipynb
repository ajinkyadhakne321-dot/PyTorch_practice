### tensor data types
**Note** tensors datatype is one of the three big errors with pytorch and deeplearning

1. tensors not right datatypes - to do get datatype from tensor
2. tensors not right shape - to do get shape from tensor
3. tensors not on the right device - to do get device from tensor



```python
#float 32 tensor
float_32_tensor = torch.tensor([3.0, 6.0, 9.0],
                               dtype=None,
                               device=None,
                               requires_grad=False)

float_32_tensor
```

```python
float_32_tensor = torch.tensor([3.0, 6.0, 9.0],
                               dtype=int)
float_32_tensor.dtype
float_32_tensor
```

```python
float_16_tensor = float_32_tensor.type(torch.float16)
float_16_tensor
```

```python
float_16_tensor * float_32_tensor
```
