### how to soleve the error with the multiplication of matrix

```python
#shapes for matrix multiplication
tensor_A = torch.tensor([[1, 2],
                         [3, 4],
                         [5, 6]])
tensor_b = torch.tensor([[1, 2],
                         [3, 4],
                         [5, 6]])
torch.mm(tensor_A, tensor_b) # torch.mm is the same as torch.matmul

```

```python
# matrix multipication workes when tensor_b is trasposed

print(f"orignal shapes: tensor_A = {tensor_A.shape}, tensor_b = {tensor_b.shape}")
print(f"new shapes: tensor_A = {tensor_A.shape}, tensor_b = {tensor_b.T.shape}")
print("multiplying:, {tensor_A.shape} @ {tensor_b.T.shape} <- inner dimension must match")
print("output:\n")
output = torch.matmul(tensor_A, tensor_b.T)
print(output)
print(f"\noutput shape: {output.shape}")
```

###reshaping, stacking, squeezing and unsqueezing
*reshaping - reshapes an input tensor to a defined shape

*view - return a view of an input tensor of certain shape but keep the same memory as the orignal tensor

*stacking - combine multiple tensor on top of each other(vstack) or side by side (hstack)

*unsqueeze - add a one dimension to a target tensor

*permute - return a view of the input with dimension permuted in a certian way

```python
import torch
x = torch.arange(1., 10.)
x, x.shape
```

```python
#add an extra dimension
x_reshaped = x.reshape(1, 9)
x_reshaped, x_reshaped.shape
```

```python
x_reshaped = x.reshape(9, 1)
x_reshaped, x_reshaped.shape
```

```python
#change the view
z = x.view(1, 9)
z, z.shape
```

```python
#changing z changes x (beacuse a view of a tensor shares the same memory as the orignal input)
z[:, 0] = 5
z, x
```

```python
#stack tensor on top of each other
x_stacked = torch.stack([x, x, x, x, x], dim = 0)
x_stacked
```

```python
x_stacked = torch.stack([x, x, x, x, x], dim = 1)
x_stacked
```

```python
#torch.squeeze() - removes all single dimensions from a target tensor
print(f"previous tensor: {x_reshaped}")
print(f"previous shape: {x_reshaped.shape}")

#remove extra dimensions from x_reshaped
x_squeezed = x_reshaped.squeeze()
print(f"\nsqueezed tensor: {x_squeezed}")
print(f"squeezed shape: {x_squeezed.shape}")
```

```python
#torch.unsqueeze() - adds a single dimension to a target tensor at a specific dim (dimension)
print(f"previous tensor: {x_squeezed}")
print(f"previous shape: {x_squeezed.shape}")
print(f"unsqueezed tensor is: {x_squeezed.unsqueeze(dim = 0)}")
print(x_squeezed.unsqueeze(dim = 0).shape)
```

```python
print(f"unsqueezed tensor is: {x_squeezed.unsqueeze(dim = 1)}")
print(x_squeezed.unsqueeze(dim = 1).shape)
```

```python
#torch.permute - rearranges the dimensions of a target tensor in specific order
x_original = torch.rand(size = (224, 224, 3)) #[height, width, colour_channels]
#permute the orignal tensor to rearrange the axis (or dim) order
x_permuted = x_original.permute(2, 0, 1) # shifts axis 0 ->1, 1 -> 2, 2 -> 0
print(f"previous dim: {x_original.shape}")
print(f"new dim: {x_permuted.shape}")
```
